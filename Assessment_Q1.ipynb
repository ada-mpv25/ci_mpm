1. **Visualise representative T1–T2 pairs**

```python
# Extract available anatomical views from the training dataset
views = train_dataset.unique("view")

def show_pairs_for_view(view_name):
    # Filter training dataset by view
    indices = [i for i, v in enumerate(train_dataset["view"]) if v == view_name]
    ds_view = train_dataset.select(indices)

    # Select 8 examples
    examples = ds_view.shuffle(seed=42).select(range(8))

    # Make the figure (8 rows × 2 columns)
    fig, axes = plt.subplots(8, 2, figsize=(3, 10))

    for i, ex in enumerate(examples):
        t1 = ex["t1"]
        t2 = ex["t2"]

        # Convert to numpy
        if isinstance(t1, torch.Tensor):
            t1 = t1.numpy()
            t2 = t2.numpy()

        # Remove channel dimension: (1, 64, 64) -> (64, 64)
        t1 = np.squeeze(t1)
        t2 = np.squeeze(t2)

        # Plot T1
        axes[i, 0].imshow(t1, cmap="gray")
        axes[i, 0].set_title(f"T1 (Example {i+1})", fontsize=8)
        axes[i, 0].axis("off")

        # Plot T2
        axes[i, 1].imshow(t2, cmap="gray")
        axes[i, 1].set_title(f"T2 (Example {i+1})", fontsize=8)
        axes[i, 1].axis("off")

    plt.suptitle(f"Anatomical view: {view_name}", fontsize=10)
    plt.tight_layout()
    plt.show()


# Generate the figures for each view
for view in views:
    show_pairs_for_view(view)
```

2. **Analyse intensity distributions**  

```python
import numpy as np
import matplotlib.pyplot as plt
import torch

# Number of slices to sample for speed
N_SAMPLE = 5000

def get_sample_pixels(dataset, N_SAMPLE, key="t1"):
    """Randomly sample N_SAMPLE slices from dataset and flatten pixels."""
    subset = dataset.shuffle(seed=42).select(range(min(N_SAMPLE, len(dataset))))
    pixels = np.concatenate([
        ex[key].numpy().squeeze().ravel()
        for ex in subset
    ])
    return pixels

# Collect pixel values
t1_train = get_sample_pixels(train_dataset, N_SAMPLE, "t1")
t2_train = get_sample_pixels(train_dataset, N_SAMPLE, "t2")
t1_valid = get_sample_pixels(valid_dataset, N_SAMPLE, "t1")
t2_valid = get_sample_pixels(valid_dataset, N_SAMPLE, "t2")

# Plot grid
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # 2 rows × 2 cols

# Row 1: Training
axes[0,0].hist(t1_train, bins=100, density=True, color='blue', alpha=0.5)
axes[0,0].set_title("Training T1")
axes[0,0].set_xlabel("Pixel intensity")
axes[0,0].set_ylabel("Density")

axes[0,1].hist(t2_train, bins=100, density=True, color='orange', alpha=0.5)
axes[0,1].set_title("Training T2")
axes[0,1].set_xlabel("Pixel intensity")
axes[0,1].set_ylabel("Density")

# Row 2: Validation
axes[1,0].hist(t1_valid, bins=100, density=True, color='cyan', alpha=0.5)
axes[1,0].set_title("Validation T1")
axes[1,0].set_xlabel("Pixel intensity")
axes[1,0].set_ylabel("Density")

axes[1,1].hist(t2_valid, bins=100, density=True, color='red', alpha=0.5)
axes[1,1].set_title("Validation T2")
axes[1,1].set_xlabel("Pixel intensity")
axes[1,1].set_ylabel("Density")

plt.tight_layout()
plt.show()
```
